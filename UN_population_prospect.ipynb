```python
japan = japan.loc[:,'Year':].rename(columns={'Year':'year','100+':100})
```

```python
japan.year = japan.year.astype(int)
```

```python
japan['lost_gen_max'] = (japan.year -1971)
japan['lost_gen_min'] = (japan.year -1982).clip(upper=100)
```

```python
japan['babyboomers_max'] = (japan.year -1945)
japan['babyboomers_min'] = (japan.year -1951).clip(upper=100)
```

```python
japan = japan.melt(id_vars=['year', 'gender','lost_gen_min', 'lost_gen_max','babyboomers_min','babyboomers_max'],var_name='age',value_name='population')
```

```python
japan['lost_gen'] = (japan.age>=japan.lost_gen_min)&(japan.age<=japan.lost_gen_max)
japan['babyboomers'] = (japan.age>=japan.babyboomers_min)&(japan.age<=japan.babyboomers_max)
```

```python
japan['generation'] = japan.lost_gen.replace({True:'Lost generation', False:'other'})
```
