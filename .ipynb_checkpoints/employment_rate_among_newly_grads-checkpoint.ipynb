```python
import pandas as pd
```

```python
url = "https://www.e-stat.go.jp/stat-search/file-download?statInfId=000031852305&fileKind=0"
```

```python
df = pd.read_excel(url, skiprows=1, header = [1,2,3], skipfooter=1,index_col=1).iloc[:,1:]
```

```python
college = df.loc[:,(df.columns.get_level_values(-1)=="計")&(df.columns.get_level_values(0)=="大学")]
```

```python
college.columns = ['employment_rate']
college.index.names = ['year']
college = college.reset_index()
```

```python
college.plot(x ='year', y="employment_rate")
```

```python
college.to_csv("./output/employment_rate.csv", index=False)
```
