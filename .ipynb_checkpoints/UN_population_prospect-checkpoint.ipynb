```python
import pandas as pd
```

```python
#both sexes
#url = "https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/EXCEL_FILES/2_Population/WPP2022_POP_F01_1_POPULATION_SINGLE_AGE_BOTH_SEXES.xlsx"
```

```python
#male
url_m = "https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/EXCEL_FILES/2_Population/WPP2022_POP_F01_2_POPULATION_SINGLE_AGE_MALE.xlsx"
#female 
url_f = "https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/EXCEL_FILES/2_Population/WPP2022_POP_F01_3_POPULATION_SINGLE_AGE_FEMALE.xlsx"
```

```python
df_m  = pd.read_excel(url_m, header=16, na_values=['*', ' '])
df_f  = pd.read_excel(url_f, header=16, na_values=['*', ' '])
```

```python
df_projection_m  = pd.read_excel(url_m, header=16, na_values=['*', ' '], sheet_name='Medium variant')
df_projection_f  = pd.read_excel(url_f, header=16, na_values=['*', ' '], sheet_name='Medium variant')
```

```python
japan_m = pd.concat([
    df_m[df_m['ISO3 Alpha-code']=='JPN'],
    df_projection_m[df_projection_m['ISO3 Alpha-code']=='JPN']
])
```

```python
japan_f = pd.concat([
    df_f[df_f['ISO3 Alpha-code']=='JPN'],
    df_projection_f[df_projection_f['ISO3 Alpha-code']=='JPN']
])
```

```python
japan_m['gender'] = 'M'
japan_f['gender'] = 'F'
```

```python
japan = pd.concat([japan_m,japan_f])
```

```python
japan
```

```python
japan = japan.loc[:,'Year':].rename(columns={'Year':'year','100+':100})
```

```python
japan.year = japan.year.astype(int)
```

```python
japan['lost_gen_max'] = (japan.year -1971)
japan['lost_gen_min'] = (japan.year -1982).clip(upper=100)
```

```python
japan['babyboomers_max'] = (japan.year -1945)
japan['babyboomers_min'] = (japan.year -1951).clip(upper=100)
```

```python
japan.head()
```

```python
japan = japan.melt(id_vars=['year', 'gender','lost_gen_min', 'lost_gen_max','babyboomers_min','babyboomers_max'],var_name='age',value_name='population')
```

```python
japan['lost_gen'] = (japan.age>=japan.lost_gen_min)&(japan.age<=japan.lost_gen_max)
japan['babyboomers'] = (japan.age>=japan.babyboomers_min)&(japan.age<=japan.babyboomers_max)
```

```python
japan['generation'] = japan.lost_gen.replace({True:'Lost generation', False:'other'})
```

```python
japan.loc[(japan.babyboomers)&(~japan.lost_gen), 'generation'] = 'Babyboomers'
```

```python
japan.population = japan.population.astype(float)
```

```python
#japan_population = japan[japan.year<=2050].drop(['lost_gen_min','lost_gen_max','babyboomers_min','babyboomers_max'], axis=1)
```

```python
japan.to_csv('./output/japan_population.csv', index=False)
```

```python
japan_aggregated = japan.pivot(index=['year','age','generation'],columns=['gender'],values='population').sum(axis=1).rename('population_in_thousands').reset_index()
```

```python
japan_aggregated['population'] = japan_aggregated.population_in_thousands *1000
```

```python
years = [2022,2040,2050, 2060]

for year in years:
    data = japan_aggregated[japan_aggregated.year==year].drop(['year'],axis=1)
    filename = f"./output/japan_population_{year}.csv"
    data.to_csv(filename, index=False)
    
    #reference data
    lostgen_population = data[data.generation=='Lost generation'].population.sum()
    lostgen_aged_population = data[(data.generation=='Lost generation')&(data.age>=65)].population.sum()
    lostgen_age = str(data[data.generation=='Lost generation'].age.min()) + '-' + str(data[data.generation=='Lost generation'].age.max())
    lostgen_pct = lostgen_population/ data.population.sum()
    aged_population = data[data.age>=65].population.sum()
    aged_pct = aged_population / data.population.sum()
    lostgen_aged_pct = lostgen_aged_population / aged_population
    print(year, ":")
    print("  lost gen=", lostgen_age, lostgen_population, str(round(lostgen_pct *100,1))+'%')
    print("  aged=", str(round(aged_pct*100,1))+"%",f", of which {lostgen_aged_pct*100:.1f}% are lost gen")
```
